```python
rape_victims2016 = "../input/crime-in-india/Victims of Rape/2016 Victims of Rape.csv"
victim_data2016 = pd.read_csv(rape_victims2016)
victim_data2016.head()
```

```python
rape_victims2017 = "../input/crime-in-india/Victims of Rape/2017 Victims of Rape.csv"
victim_data2017 = pd.read_csv(rape_victims2017)
victim_data2017.head()
```

```python
rape_victims2018 = "../input/crime-in-india/Victims of Rape/2018 Victims of Rape.csv"
victim_data2018 = pd.read_csv(rape_victims2018)
victim_data2018.head()
```

```python
plt.figure(figsize = (20, 8))
plt.title("The number of  child rape cases reported in 2016 from each state in India")
sns.barplot( x = victim_data2016['Child Victims of Rape (Below 18 Yrs) - Below 6 Years'], y = victim_data2016['Cases Reported'])
```

```python
plt.figure(figsize = (20, 8))
plt.title("The number of  child rape cases reported in 2017 from each state in India")
sns.barplot( x = victim_data2017['Child Victims of Rape (Below 18 Yrs) - Below 6 Years'], y = victim_data2017['Cases Reported'])
```
